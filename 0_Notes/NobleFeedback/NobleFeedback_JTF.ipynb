```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('3_Day_NBL.csv');
```

```python
def create_bar_chart(pd_series):
    return (pd_series.astype('str')
                     .str.split(',', expand = True)
                     .apply(pd.Series.value_counts)
                     .sum(axis=1)
                     .sort_values()
                     .plot(kind='barh', figsize=(8,8), color='darkgreen', title=pd_series.name))
```

```python
for column in df.iloc[:, 9:]:
    plt.show(create_bar_chart(df[column]))
    print("   ")
```
