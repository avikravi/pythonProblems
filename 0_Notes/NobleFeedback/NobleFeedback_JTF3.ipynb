```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('3_Day_NBL.csv');
```

```python
def create_bar_chart(pd_series, save_figure=False):
    (pd_series.astype('str')
                     .str.split(',', expand=True)
                     .apply(pd.Series.value_counts)
                     .sum(axis=1)
                     .sort_values()
                     .plot(kind='barh', figsize=(8,8), color='darkgreen', title=pd_series.name))
    
    
    if save_figure:
        filepath = '{}.png'.format(pd_series.name.replace('/','-'))
        plt.savefig(filepath)
    
    plt.show()
```

```python
df.iloc[:, 9:].apply(create_bar_chart, args=(True,))
```
