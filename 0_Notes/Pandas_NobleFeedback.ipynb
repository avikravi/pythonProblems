```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('3_Day_NBL.csv')
```

```python
df['What pre-requisites would be most helpful to participants? (Check all that apply)'].value_counts().plot(kind='barh',
                                    figsize=(8,8), color='darkgreen')
```

```python
df['Which group / job function would benefit from taking Energy programming / analytics / data science courses? (Check all that apply)'].value_counts().plot(kind='barh',
                                    figsize=(8,8), color='darkgreen')
```

```python
df['How would you improve this workshop? (check all that apply)'].value_counts().plot(kind='barh',
                                    figsize=(8,8), color='darkgreen')
```

```python
df['Which other platforms would you like to learn Machine Learning / Data Analytics / Data Science (ML-DADS) on? (Check all that apply)'].value_counts().plot(kind='barh',
                                    figsize=(8,8), color='darkgreen')
```
